```python
from sklearn.datasets import load_boston
```

```python
boston = load_boston()
```

```python
print = boston.DESCR
```

```python
plt.hist(boston.target,bins=50)
plt.xlabel('Price in $1000')
plt.ylabel('Number of Houses')
```

```python
boston_df['Price'] = boston.target
```

```python
sns.lmplot('RM','Price',data=boston_df)
```
